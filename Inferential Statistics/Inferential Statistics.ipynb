# Problem Statement 2

#### The breaking strength of gunny bags used for packaging cement is normally distributed with a mean of 5 kg per sq. centimeter and a standard deviation of 1.5 kg per sq. centimeter. The quality team of the cement company wants to know the following about the packaging material to better understand wastage or pilferage within the supply chain; Answer the questions below based on the given information; (Provide an appropriate visual representation of your answers, without which marks will be deducted)

## 2.1 What proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm?

#### Given: Mean= 5, Standard Deviation = 1.5

```python
z1=(3.17-5)/1.5
z11=stats.norm.cdf(z1)
print('The Z score is ',z1)
print('There is ',round((z11*100),2), '% chance that the gunny bags have a breaking strength less than 3.17 kg per sq cm') 
```

```python
x=np.linspace(-3,3,1000)
pdf = norm.pdf(x)
plt.figure(figsize=(8, 4))
plt.plot(x, pdf,'b-')
plt.fill_between(x, pdf, where=(x <= z1), alpha=0.3, color='r');
plt.xlabel('Distribution Range');
plt.title('Proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm');
```

#### Fig: 1.2 - Proportion of the gunny bags have a breaking strength less than 3.17 kg per sq cm

## 2.2 What proportion of the gunny bags have a breaking strength at least 3.6 kg per sq cm.?

```python
z2=(3.5-5)/1.5
z22=1-stats.norm.cdf(z2)
print('The Z score is ',z2)
print('There is ',round((z22*100),2), '% chance that the gunny bags have a breaking strength at least 3.6 kg per sq cm') 
```

```python
x=np.linspace(-3,3,1000)
pdf = norm.pdf(x)
plt.figure(figsize=(8, 4))
plt.plot(x, pdf,'b-')
plt.fill_between(x, pdf, where=(x > z2), alpha=0.3, color='g');
plt.xlabel('Distribution Range');
plt.title('Proportion of the gunny bags have a breaking strength at least 3.6 kg per sq cm');
```

#### Fig: 1.3 - Proportion of the gunny bags have a breaking strength at least 3.6 kg per sq cm

## 2.3 What proportion of the gunny bags have a breaking strength between 5 and 5.5 kg per sq cm.?

```python
z3=(5-5)/1.5
z4=(5.5-5)/1.5
z34=stats.norm.cdf(z4)-stats.norm.cdf(z3)
print('The Z score of 5kg per sq cm is ',z3)
print('The Z score is 5.5kg per sq cm ',z4)
print('There is ',round((z34*100),2), '% chance that the gunny bags have a breaking strength between 5 and 5.5kg per sq cm') 
```

```python
x=np.linspace(-3,3,1000)
pdf = norm.pdf(x,0,1)
plt.figure(figsize=(8, 4))
plt.plot(x,pdf,'r-')
plt.fill_between(x,0,pdf, where=(x >= z3 ) & (x <= z4 ), color='blue', alpha=0.3);
plt.xlabel('Distribution Range');
plt.title('Proportion of the gunny bags have a breaking strength between 5 and 5.5kg per sq cm');
```

#### Fig: 1.4 - Proportion of the gunny bags have a breaking strength between 5 and 5.5kg per sq cm

## 2.4 What proportion of the gunny bags have a breaking strength NOT between 3 and 7.5 kg per sq cm.?

```python
z5=(3-5)/1.5
z55=stats.norm.cdf(z5)
z6=(7.5-5)/1.5
z66=1-stats.norm.cdf(z6)
z56=z55+z66
print('The Z score of less than 3kg per sq cm is ',z5)
print('The Z score is greater than 7.5kg per sq cm ',z6)
print('There is ',round((z56*100),2), '% chance that the gunny bags have a breaking strength NOT between 3 and 7.5kg per sq cm')
```

```python
x=np.linspace(-3,3,1000)
pdf = norm.pdf(x,0,1)
plt.figure(figsize=(8, 4))
plt.plot(x,pdf,'r-')
plt.fill_between(x,0,pdf, where=(x < z5 ), color='b', alpha=0.5);
plt.fill_between(x,0,pdf, where=(x > z6 ), color='b', alpha=0.5);
plt.xlabel('Distribution Range');
plt.title('Proportion of the gunny bags have a breaking strength NOT between 3 and 7.5kg per sq cm');
```

## 3.1 Zingaro has reason to believe that the unpolished stones may not be suitable for printing. Do you think Zingaro is justified in thinking so?

```python
df_zingaro=pd.read_csv('Zingaro_Company.csv')
```

### 3.1.1 State the null and alternate hypothesis 

#### Ho: The unpolished stones may be suitable for printing (where the stone surface has Brinell's hardness index >=150)
#### Ha: The unpolished stones may not be suitable for printing (where the stone surface has Brinell's hardness index <150)
