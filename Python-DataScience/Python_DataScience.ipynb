## Q 5. Find the Total Revenue generated by the E-Commerce company over all orders placed. (6 marks)

#### Hint: Calculate revenue of E-commerce company using the conditions mentioned above and then do the total (sum) of all to get total Revenue. 

### Ans 5

##### Ans 5 - 
1. If else conditions are used to executive the conditions given
2. for loop is used to itirate through the coulmn of discounted price
3. append function is used to added the computed output of if else function
4. The computed output is stored in list called revenue
5. the list revenue is added in the dataset by assigning new column name 'revenue'
6. The total revenue is calculated using sum() function for the E-Commerce company

```python
revenue = []
for i in df['discounted_price']:
    if i > 600:
        revenue.append((i*0.25))
    elif i >350  and i <= 600:
        revenue.append((i*0.15))
    elif i >100  and i <= 350:
        revenue.append((i*0.10))    
    else:
        revenue.append((i*0.05))
```

```python
df["revenue"] = revenue
df
```

```python
a=df['revenue'].sum()
print('The total revenue is ',a)
```

#### The total revenue is  2217486.85

## Q 9.Compare performance regionwise

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

### Ans 9 a)

##### Ans 9 a) - 

1. Add a new column to dataset named date and change the same to datetime format using to.datetime function 
2. Add a new column to dataset named month using .dt.month function of the column 'date'
3. line plot function to create a line for 4 regions separately with hue as region

```python
df['date']  = pd.to_datetime(df['Order_Date'],dayfirst=True)
df.info()
```

```python
df['Month'] = df['date'].dt.strftime('%m-%y')
df
```

```python
plt.figure(figsize = (30,7))
g = sns.FacetGrid(df, col="Region", hue="Region", col_wrap=5)
plt.xticks(rotation=90)
g.map(sns.lineplot, "Month","revenue",errorbar=('ci',False))
plt.show()
```
