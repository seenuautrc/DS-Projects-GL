### 2.1.2 Outlier Treatment

```python
def remove_outlier(col):
    Q1,Q3=np.percentile(col,[25,75])
    IQR=Q3-Q1
    LL=Q1-(1.5*IQR)
    UL=Q3+(1.5*IQR)
    return LL,UL

lr,ur=remove_outlier(df['No_of_Dependents'])
print('Lower Range',lr,'Upper Range',ur)
```

```python
df['No_of_Dependents']=np.where(df['No_of_Dependents']<lr,lr,df['No_of_Dependents'])
df['No_of_Dependents']=np.where(df['No_of_Dependents']>ur,ur,df['No_of_Dependents'])
```

```python
sns.boxplot(data=df,x='No_of_Dependents',color='black');
```

```python
# The outliers are caped to their Lower Limit/Lower Range
```

### 2.2.3 Numerical vs Categorical Variable Analysis

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define the columns you want to create box plots for
numerical_columns = ['Age', 'No_of_Dependents', 'Salary', 'Partner_salary', 'Total_salary', 'Price']
categorical_columns = ['Gender', 'Profession', 'Marital_status', 'Education', 'Personal_loan', 'House_loan', 'Partner_working', 'Make']

# Create subplots
fig, axes = plt.subplots(len(numerical_columns), len(categorical_columns), figsize=(30,20))

# Nested loop to generate box plots for each combination
for i, num_col in enumerate(numerical_columns):
    for j, cat_col in enumerate(categorical_columns):
        ax = axes[i, j]
        sns.boxplot(y=df[num_col], x=df[cat_col], ax=ax)
        ax.set_title(f'{num_col} vs {cat_col}')
        ax.set_xlabel(cat_col)
        ax.set_ylabel(num_col)

# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
